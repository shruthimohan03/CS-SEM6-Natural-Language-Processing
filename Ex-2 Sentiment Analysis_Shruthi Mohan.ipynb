### Task 2:Perform cleaning and preprocessing of text. 

```python
import pandas as pd
train_df = pd.read_csv(r'train.csv', delimiter=',', encoding='ISO-8859-1')
train_df = train_df.iloc[:5000,:]
train_df
```

# TEXT PREPROCESSING

```python
train_df['text'].fillna('',inplace=True)
```

```python
train_df.isnull().sum()
```

### Numeric encoding

```python
def func(sentiment):
    if sentiment =='positive':
        return 0;
    elif sentiment =='negative':
        return 1;
    else:
        return 2;
train_df['sentiment'] = train_df['sentiment'].apply(func)
```

### Convert text to lower case

```python
train_df['clean_text'] = train_df['text'].str.lower()
train_df.head()
```

### Removing punctuations

```python
import re

# Variable to replace all characters that are not letters or whitespace
regex = re.compile('[^a-z1-9\s]')
# Removes all characters that are not letters or spaces
train_df['clean_text'] = train_df['clean_text'].apply(lambda x: regex.sub('', x))
train_df.head()
```

### Tokenization

```python
import nltk
#nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize
train_df['tokens'] = train_df['clean_text'].apply(word_tokenize)
train_df.head()
```

### Removing Stop Words

```python
#nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
train_df['tokens'] = train_df['tokens'].apply(lambda x: [word for word in x if word not in stop_words])
train_df.head(20)
```

### Word Normalization

```python
#nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
train_df['tokens'] = train_df['tokens'].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])
train_df.head()
```

### TASK 3: Generate representations

# 1) BAG OF WORDS EMBEDDING

### Creating the frequency dataset

```python
dataset = {} 
for words_list in train_df['tokens']:
    for word in words_list: 
        if word not in dataset.keys(): 
            dataset[word] = 1
        else: 
            dataset[word] += 1
dataset
```

```python
len(dataset)
```

##### Since the corpus size is huge, we will take the first 100 frequent words alone to build the BOW model

```python
import heapq 
freq_words = heapq.nlargest(100, dataset, key=dataset.get)
```

```python
freq_words
```

### Building the Bag of Words model
##### In this step we construct a vector, which would tell us whether a word in each sentence is a frequent word or not. 
##### If a word in a sentence is a frequent word, we set it as 1, else we set it as 0.

```python
import numpy as np
bow = [] 
for data in train_df['clean_text']: 
    vector = [] 
    for word in freq_words: 
        if word in data: 
            vector.append(1) 
        else: 
            vector.append(0) 
    bow.append(vector) 
bow = np.asarray(bow) 
bow
```

# 3) Continuous Bag of Words

### Convert the corpus to a sequence of integers

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer()
tokenizer.fit_on_texts(corpus)
sequences = tokenizer.texts_to_sequences(corpus)
print("After converting our words in the corpus \
into vector of integers:")
print(sequences)
```

### Building the CBOW model having window size = 2 (no of context words considered)

```python
# Define the parameters
vocab_size = len(tokenizer.word_index) + 1
embedding_size = int(1.6*np.sqrt(len(dataset))) #embedding size=what dimension should each object be represented in
#how to choose embedding size= 1.6*sqrt(unique features)
print('embedding size:',embedding_size)
window_size = 2
```

```python
# Generate the context-target pairs
contexts = []
targets = []
for sequence in sequences:
    for i in range(window_size, len(sequence) - window_size):
        context = sequence[i - window_size:i] + sequence[i + 1:i + window_size + 1]
        target = sequence[i]
        contexts.append(context)
        targets.append(target)

# Convert the contexts to numpy array
contexts = np.array(contexts)
contexts
```

### Defining the CBOW model

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Embedding, Lambda
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
model = Sequential()
model.add(Embedding(input_dim=vocab_size,output_dim=embedding_size,input_length=2*window_size))
model.add(Lambda(lambda x: tf.reduce_mean(x, axis=1)))
model.add(Dense(units=vocab_size, activation='softmax'))
```

```python
# Get the word embeddings
embeddings = model.get_weights()[0]

# Perform PCA to reduce the dimensionality of the embeddings
pca = PCA(n_components=2)
reduced_embeddings = pca.fit_transform(embeddings)
```
